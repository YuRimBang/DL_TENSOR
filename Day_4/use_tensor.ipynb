### 텐서 저장 및 실행 실습

```python
import tensorflow as tf
from MyUtils import *
```

```python
### 임의의 숫자 데이터를 (10,20) 텐서로 저장
### 조건 : 숫자 데이터의 분포가 정규분포(평균 0, 표준 편차 1)
### 위치 : GPU가 있다면 GPU에서 저장 및 실행

# 매번 동일한 임의의 숫자 추출을 위한 고정 -> 학습이 안정된 후 풀기
tf.random.set_seed(5)

# 저장 및 실행 위치 설정
DEVICE = getDeviceName()
```

```python
with tf.device(DEVICE):
    numTS = tf.random.normal([10,20])
    print(f'{numTS}\n{numTS}')
```

- 변경 가능 데이터 텐서 --> tf.Variable()

```python
# 나이 데이터 저장
ageTS = tf.Variable(23)

print(f'ageTS   -> {type(ageTS)}')
print(f'shape   -> {ageTS.shape}')
print(f'dtype   -> {ageTS.dtype}')
print(f'{ageTS}')
print(f'numpy   -> {ageTS.numpy()}')

#ResourceVariable
```

```python
# 나이 데이터 변경 
ageTS.assign(24)

print(ageTS)
```

- 변경 불가 데이터 텐서 -> tf.constant()

```python
# 국가 코드 데이터 저장 -> 대한민국(82)
loc_code = tf.constant(82)

print(f'loc_code-> {type(loc_code)}')
print(f'shape   -> {loc_code.shape}')
print(f'dtype   -> {loc_code.dtype}')
print(f'ndim    -> {loc_code.ndim}D')
print(f'numpy   -> {loc_code.numpy()}')

#EagerTensor
```

```python
# 국가 코드 변경
loc_code.assign(100)

#AttributeError: 'tensorflow.python.framework.ops.EagerTensor' object has no attribute 'assign'
```
